```python
data.info()
```

```python
import matplotlib.pyplot as plt
box=data.boxplot(figsize=(16,6), )
box.set_xticklabels(data.columns,rotation=30)           
```

```python
from sklearn.cluster import KMeans

err=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(data)
    err.append(kmeans.inertia_)
plt.plot(range(1,11),err)
plt.title('Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('error')
plt.show
```

```python
kmeans = KMeans(n_clusters=6)
data_labels = kmeans.fit_predict(data)
data_labels
```

```python
data['clusters'] = data_labels
```

```python
data.head()
```

```python
data1 = data[data.clusters == 0]
data2 = data[data.clusters == 1]
data3 = data[data.clusters == 2]
data4 = data[data.clusters == 3]
data5 = data[data.clusters == 4]
data6 = data[data.clusters == 5]
```

```python
f=['venue.location.lat','venue.location.lng']
kmeans_parameters = nearby_venues[f]
```

```python
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
```

```python
error=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, max_iter=300)
    kmeans.fit(kmeans_parameters)
    ##kmeans.fit(nearby_venues['venue.location.lat', 'venue.location.lat'])
    error.append(kmeans.inertia_)

k = [i*100 for i in np.diff(error,2)].index(min([i*100 for i 
     in np.diff(error,2)]))


plt.plot(range(1,11),error)
plt.axvline(k, ls='--', color="red", label="k = "+str(k))
plt.title('Elbow Method')
plt.xlabel('No. of clusters')
plt.ylabel('error')
plt.show
```
